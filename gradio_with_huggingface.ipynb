```python
import transformers
```

```python
from diffusers import DiffusionPipeline
import torch

pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, use_safetensors=True, variant="fp16")
pipe.to("cuda")

```

```python
from transformers import pipeline

captioner = pipeline("image-to-text",model="Salesforce/blip-image-captioning-base")

## [{'generated_text': 'two birds are standing next to each other '}]
```

```python
import cv2
def textimage(inp):
  images = pipe(inp).images[0]


  return images


def imagetext(img2):
  cv2.imwrite("img1.jpg",img2)
  txt=captioner("/content/img1.jpg")

  return  txt

```

```python
!pip install gradio
```

```python
import gradio as gr
wp=gr.Interface(fn=textimage, inputs=["text","image"],outputs=["image","text"],title="AI class")
wp.launch(share=True,auth=("raheem","ai"))
```

```python
with gr.Blocks() as dm:
  with gr.Row():
    inp=gr.Textbox()
    out=gr.Image()
    inp2=gr.Image()
    out2=gr.Textbox()

  btn = gr.Button("Run1")
  btn.click(fn=textimage, inputs=inp, outputs=out)
  btn = gr.Button("Run2")
  btn.click(fn=imagetext, inputs=inp2, outputs=out2)
dm.launch(share=True)
```
