```python
import tensorflow as tf
from tensorflow import keras
model=keras.models.load_model("cnnmodel.h5")
import cv2
import numpy as np
import pandas as pd
```

```python
cap=cv2.VideoCapture(0)
while True:
    b,img=cap.read()
    im=cv2.resize(img,(64,64))
    ip=np.expand_dims(im,axis=0)
    pred=model.predict(ip)
    pre=np.argmax(pred,axis=1)
    if pre==0:
        cv2.putText(img,"raheem",(100,100),3,2,(0,255,0))
    if pre==1:
        cv2.putText(img,"mobile",(100,100),3,2,(255,0,0))
    cv2.imshow("raheem",img)
    cv2.waitKey(1)
    
```
